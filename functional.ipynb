````python
# Produce markdown presentation of a table for simple debugging

def markdown_table(table: list[dict[str, str | float]]) -> str:
    '''Ouput table in markdown format'''

    column_headers = table[0].keys()
    num_columns = len(column_headers)

    header = f"| {' | '.join(column_headers)} |"
    header_separator = f"{'|------' * num_columns}|"

    # Loop through the list of dictionaries representing the table rows
    markdown_rows = [
        f"| {' | '.join(str(val) for val in row.values())} |"  # A single markdown row
        for row in table
    ]
    str_all_mardown_rows = '\n'.join(markdown_rows)

    return f'{header}\n{header_separator}\n{str_all_mardown_rows}'
````
